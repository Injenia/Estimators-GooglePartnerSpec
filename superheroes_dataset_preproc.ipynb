```python
from tensorflow.python.lib.io.file_io import FileIO as open_file
import pandas as pd
import json
import math
```

```python
DATASET_CSV="gs://ml-research-injenia/estimators/datasets/superhero-set/joined.csv"
OUT_DIR="gs://ml-research-injenia/estimators/datasets/superhero-set"
with open_file(DATASET_CSV, "r") as f:
    original_df=pd.read_csv(f)
original_df
```

```python
df=original_df.copy()
df.columns=[x.strip().replace(" ", "_").replace("_-_", "_") for x in df.columns]
df=df[df["Alignment"].isin(['good', 'bad', 'neutral'])]
```

```python
for c in df.columns:
    if df[c].dtypes.name == 'bool':
        df[c]=df[c].map({True: 'TRUE', False: 'FALSE'})
df
```

```python
LABEL_COLUMN="Alignment"
eval_percent=0.1
```

```python
dataset_df=df
with open_file(OUT_DIR+"/preproc_dataset.csv","w") as f:
    dataset_df.to_csv(f,index=False)
```

```python
labels=dataset_df[LABEL_COLUMN].unique()
labels
```

```python
individuals_df={l:dataset_df[dataset_df[LABEL_COLUMN]==l].sample(frac=1) for l in labels}
splitted_dfs={}
for l in labels:
    splitted_dfs[l]={
        "eval":individuals_df[l][:int(float(individuals_df[l].shape[0])*eval_percent)],
        "train":individuals_df[l][int(float(individuals_df[l].shape[0])*eval_percent):]
    }
train_set=pd.concat([splitted_dfs[l]["train"] for l in splitted_dfs]).sample(frac=1)
eval_set=pd.concat([splitted_dfs[l]["eval"] for l in splitted_dfs]).sample(frac=1)
```

```python
with open_file(OUT_DIR+"/preproc_trainset.csv","w") as f:
    train_set.to_csv(f,index=False)
with open_file(OUT_DIR+"/preproc_evalset.csv","w") as f:
    eval_set.to_csv(f,index=False)
```
